```python
# data
df = sns.load_dataset('titanic')
```

```python
df.info()
```

```python
# Scatter Plot
```

```python
# It is Build through 2 Numeric Variables
# figure size
plt.figure(figsize=(10,8)) # 10,8 is A4 size paper, We can Make Plot Accordingly to Slides Size etc
sns.scatterplot(data=df, x='age',y='fare',hue='sex')
```

```python
# It is Build through 2 Numeric Variables
# figure size
#plt.figure(figsize=(10,8)) # 10,8 is A4 size paper, We can Make Plot Accordingly to Slides Size etc
#sns.scatterplot(data=df, x='age',y='fare',hue='sex')

# Change the Scatter Point Color
palette = { "male" : "red", "female" : "green"}
sns.scatterplot(data=df, x='age',y='fare',hue='sex', palette=palette)
```

```python
# Saving Plot
plt.savefig('scatter_plot.png',dpi=300)
```

```python
# Style of a Plot
plt.figure(figsize=(10,8))
sns.set_theme(style='darkgrid',font_scale=1.5)
palette = { "male" : "red", "female" : "green"}
sns.scatterplot(data=df, x='age',y='fare',hue='sex', size="fare",palette=palette)
```


# `Interpretation:`

1.We have Age of passengers at x-axis and fare at y-axis.

2.The scatter plot shows the relationship between age and fare of Titanic passengers.

3.We further grouped the passengers by their genders such as male (Green points) and female (Red points).

4.Two male passengers gave the highest fare in Titanic, and they might be traveling first class.

5.The oldest person on Titanic is also a male having 80 years of age as shown in the scatter plot. (As shown in Figure 1).

```python
# Style of a Plot
plt.figure(figsize=(12,6))
sns.set_theme(style='darkgrid',font_scale=1)
palette = { "First" : "red", "Second" : "green", "Third" : "yellow"}
sns.boxplot(data=df, x='sex',y='age',hue='class', palette=palette)

# Change Legend Title and Location
plt.legend(loc='lower right',title='Class Table')
```

```python
# Style of a Plot
plt.figure(figsize=(12,6))
sns.set_theme(style='darkgrid',font_scale=1)
palette = { "First" : "red", "Second" : "green", "Third" : "yellow"}
sns.boxenplot(data=df, x='sex',y='age',hue='class', palette=palette)
plt.xlabel("Gender")
plt.ylabel("Age")

plt.xticks(ticks=[0,1],labels={'Male','Female'},rotation=0)

# Change Legend Title and Location
plt.legend(loc='lower right',title='Class Table')
```

```python
# Make Scatter Plot in Plotly
import plotly.express as px

fig = px.scatter(df, x="age", y="fare",color="sex")
fig.show()
```

```python
# Bar Plot Using Seaborn
# Style of a Plot
plt.figure(figsize=(12,6))
sns.set_theme(style='darkgrid',font_scale=1)
palette = { "First" : "red", "Second" : "green", "Third" : "yellow"}
sns.barplot(data=df, x='sex',y='age',hue='class', palette=palette,capsize=0.5)
plt.xlabel("Gender")
plt.ylabel("Age")

plt.xticks(ticks=[0,1],labels={'Male','Female'},rotation=0)

# Change Legend Title and Location
plt.legend(loc='lower right',title='Class Table')
```

```python
plt.figure(figsize=(12,6))
sns.set_theme(style='darkgrid',font_scale=1)
palette = { "First" : "red", "Second" : "green", "Third" : "yellow"}
sns.violinplot(data=df, x='sex',y='age',hue='class', palette=palette)
plt.xlabel("Gender")
plt.ylabel("Age")

plt.xticks(ticks=[0,1],labels={'Male','Female'},rotation=0)

# Change Legend Title and Location
plt.legend(loc='lower right',title='Class Table')
```
