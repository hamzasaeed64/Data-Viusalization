```python
import plotly.express as px
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'names': ['Alice', 'Bob', 'Charlie'], 'height': [5.8, 5.9, 5.7]}

# Create DataFrame
df = pd.DataFrame(data)

# Create Line Plot
fig = px.bar(df, x='names', y='height')
fig
```

```python
# Plotting on Titanic Dataset
df = sns.load_dataset('titanic')
df.head()
```

```python
df.dropna(inplace=True)
# Plotting on Titanic Dataset through Plotly
fig = px.scatter(df, x='age',y='fare',color="sex",size="fare",title='Plot on Titanic Dataset')
```

```python
fig.show()
```

```python
# Update the Plot
fig.update_layout(
    title='Scatter Plot',


)
```

```python
# Correlation Matrix Heatmap
px.imshow(df[['age','fare']].corr())
```

```python
df[['age','fare']].corr()
```

```python
iris = sns.load_dataset('iris')
iris.columns
```

```python
numeric_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
correlation_matrix = iris[numeric_cols].corr()

# Plot heatmap
fig = px.imshow(correlation_matrix, text_auto=True, color_continuous_scale='RdBu_r')
fig.show()
```
